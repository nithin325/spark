```python
from sklearn.preprocessing import StandardScaler
```

```python
sc = StandardScaler()
```

```python
sc.fit(df1)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```python
x = df1.drop(labels='price',axis=1)
y = df1['price']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=24)
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train,y_train)
print(f'R^2 score for train: {lr.score(x_train, y_train)}')
print(f'R^2 score for test: {lr.score(x_test, y_test)}')
```

```python
import statsmodels.api as sm

X_constant = sm.add_constant(x)
model = sm.OLS(y,X_constant).fit()
model.summary()
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(52)
best_features = pca.fit(x)
best_features.explained_variance_
x_pca = best_features.transform(x)
```

```python
x_pca_df = pd.DataFrame(x_pca)
x_pca_df.shape
```

```python
x_pca_df.head()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_pca_df,y,test_size=0.2,random_state=24)
```

```python
model = lr.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
